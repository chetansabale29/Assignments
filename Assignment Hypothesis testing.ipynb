```python
#Q-2
data = pd.read_csv('G:\CHETAN\Assignments\Hypothesis Testing\LabTAT.csv')
data
```

```python
#Normality test
#H0=lab1,lab2,lab3,lab4 are normal
#H1=lab1,lab2,lab3,lab4 are not normal
p_val1=stats.shapiro(data.iloc[:,0])
print(p_val1)
p_val2=stats.shapiro(data.iloc[:,1])
print(p_val2)
p_val3=stats.shapiro(data.iloc[:,2])
print(p_val3)
p_val4=stats.shapiro(data.iloc[:,3])
print(p_val4)
```

```python
#p_val>0.05
#accept null hypothesis = lab1,lab2,lab3,lab4 are normal
```

```python
#Variance test
#H0=Var of 4 lab is same
#H1 =var of 4 lab is not same
p_val=stats.levene(data.iloc[:,0],data.iloc[:,1],data.iloc[:,2],data.iloc[:,3])
p_val
```

```python
#p_val>0.05
#accept null hypothesis = Var of 4 lab is same
```

```python
#H0 = there is no difference bet average of lab TAT
#H1 = there is difference bet average of lab TAT
# Column are more so we use Anova f-test
p_val=stats.f_oneway(data.iloc[:,0],data.iloc[:,1],data.iloc[:,2],data.iloc[:,3])
p_val
```

```python
#pvalue<0.05
#there is difference bet average of lab TAT
```
